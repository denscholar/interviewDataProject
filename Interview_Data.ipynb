## Scrapping a website

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup
# html = urlopen('http://danielkummer.github.io/git-flow-cheatsheet/')

def getTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try:
        bsObj = BeautifulSoup(html.read())
        title = bsObj.body.h1
    except AttributeError as e:
        return None
    return title

title = getTitle('http://danielkummer.github.io/git-flow-cheatsheet/')
if title == None:
    print('Title could not be found')
else:
    print(title)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen('https://likeparadise.com.ng/phones_accessories')
bsObj = BeautifulSoup(html, 'lxml')
nameList = bsObj.find_all('span', {'class': 'price-new'})
for name in nameList:
    print(name.get_text())
```
